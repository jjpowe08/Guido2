```python
# Create a function to define those who are highly educated vs not
def grad_univ_hs_other(education):
    education, sex = education
    
    if education < 3 :
        return 'HEduc'
    else:
        return 'LEduc'
```

```python
credit['edtype'] = credit[['EDUCATION', 'SEX']].apply(grad_univ_hs_other, axis=1)
```
